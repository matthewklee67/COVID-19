```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import json
%matplotlib inline 

#Read in data
covidtesting = pd.read_csv("covidtesting.csv")
vaccine_doses = pd.read_csv("vaccine_doses.csv")

dates = covidtesting.loc[:,"Reported Date"].values
dates = pd.to_datetime(dates)
dailyPositives = covidtesting.loc[:,"Confirmed Positive"].values
totalCases = covidtesting.loc[:,"Total Cases"].values

vaccineDates = vaccine_doses.loc[:,"report_date"].values
vaccineDates = pd.to_datetime(vaccineDates)
dailyDosesAdministered = vaccine_doses.loc[:,"previous_day_doses_administered"].values
totalDoses = vaccine_doses.loc[:,"total_doses_administered"].values
totalDosesCompleted = vaccine_doses.loc[:,"total_vaccinations_completed"].values
```

```python
plt.plot(dates, dailyPositives, label = "COVID-19 Positive Cases")
plt.plot(vaccineDates, dailyDosesAdministered, label = "COVID-19 Vaccinations")
plt.xticks(rotation=90)
plt.title("COVID-19 Daily Positives vs Daily Vaccine Doses")
plt.ylabel("Amount")
plt.xlabel("Date")
plt.legend()
```

```python
plt.plot(dates, totalCases, label = "COVID-19 Positive Cases")
plt.plot(vaccineDates, totalDoses, label = "COVID-19 Vaccinations")
plt.xticks(rotation=90)
plt.title("COVID-19 Total Positives vs Total Vaccine Doses")
plt.ylabel("Amount")
plt.xlabel("Date")
plt.legend()
```
